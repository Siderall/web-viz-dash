```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
file_to_load = "Resources/cities.csv"

cities_data = pd.read_csv(file_to_load)
cities_data
```

```python
cities_data.to_html("Resources/cities.html")
```

```python
plt.scatter(cities_data["Lat"], cities_data["Max Temp"], num=None, figsize=(8, 6), dpi=80, marker="o", facecolor="purple", edgecolors="black")
plt.title("CITIES: How does Latitude affect Temperature?")
plt.ylabel("Max Temperature in F")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig("Resources/test.png")

plt.show()
```
